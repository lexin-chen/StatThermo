A. Given a number N, calculate N! using two different methods:

B. Using one of the methods derived above, compute and plot N! vs N, with axes labelled. 

```python
from matplotlib import pyplot as plt
import numpy as np
x = np.arange(0, 171, dtype='int')
x1 = list(x)
y = []
for i in x:
  j = np.math.factorial(i)
  y.append(j)
plt.xlabel("N")
plt.ylabel("N! (*1e304)")
plt.plot(x1, y)
plt.show()
```

F. Create a plot showing ln N! vs N computed directly (part D) and approximated via part E. 
Add labels for the different lines. 

```python
from matplotlib import pyplot as plt
import numpy as np
import math
from numpy import log as ln
x = np.arange(2, 171, dtype='int')
x_value = list(x)
noprox = []
for i in x_value:
  j = ln(float(np.math.factorial(i)))
  noprox.append(j)
yesprox = []
for i in x_value:
  k = i * ln(i) - i
  yesprox.append(k)
print(noprox)
print(yesprox)
plt.xlabel("N")
plt.ylabel("ln(N!)")
plt.plot(x_value, noprox, color='pink', label="ln(N!)", linewidth=5.0)
plt.plot(x_value, yesprox, color='black', label="Stirling's Approximation")
plt.legend()
plt.show()
```

G. Create a plot showing (ln N!  Stirling_approx(N!)/ ln N! to look at the relative error when 
using the approximation. 

```python
x_set = np.arange(2, 171, dtype='int')

original = [] 
approx = []
for p in x_set:
  j = ln(float(np.math.factorial(p)))
  original.append(j)
for p in x_set:
  k = p * ln(p) - p
  approx.append(k)
aa = np.array(original)
bb = np.array(approx)
cc = []
cc = abs(aa - bb) / aa


plt.xlabel("N")
plt.ylabel("Relative Error")
plt.plot(x_set, cc, color='mediumslateblue')
plt.show()
```

A) Plot the probability density for a normalized Gaussian Distribution with average zero and
standard deviation = 1

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mu = 0
sigma = .1
sigma1 = .01
sigma2 = .001
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
x1 = np.linspace(mu - 3 * sigma1, mu + 3 * sigma1, 100)
x2 = np.linspace(mu - 3 * sigma2, mu + 3 * sigma2, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma), label="sigma=0.1")
plt.plot(x1, stats.norm.pdf(x1, mu, sigma1), label="sigma=0.01")
plt.plot(x2, stats.norm.pdf(x2, mu, sigma2), label="sigma=0.001")
plt.legend()
plt.ylim(0)
plt.show()
```
